```python
df = df.sort_values(by=['HomeTeam', 'Date'], ascending=True).reset_index()
df
```

```python
df['opp_team'] = df['AwayTeam'].astype('category').cat.codes
df['home_team'] = df['HomeTeam'].astype('category').cat.codes
df
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
rf = RandomForestClassifier(criterion='entropy', n_estimators=50, min_samples_split=10, random_state=1)
train = df.loc[df['Date'] < '2020-01-01']
test = df.loc[df['Date'] > '2020-01-01']
predictor = ['home_team', 'opp_team', 'day', 'month', 'year']
rf.fit(train[predictor], train['final_result'])
```

```python
prediction = rf.predict(test[predictor])
acc = metrics.accuracy_score(test['final_result'], prediction)
```

```python
team_group = df.groupby('HomeTeam')
group = team_group.get_group('Man United').sort_values('Date')
group
```

```python
def rolling_averages (group, colms, new_col):
    group = group.sort_values("Date")
    roll_stats = group[colms].rolling(3, closed='left').mean()
    group[new_col] = roll_stats
    group = group.dropna(subset=new_col)
    return group
```

```python
colms = ['FTHG', 'FTAG','HTHG','HTAG', 'HS', 'AS', 'HST','AST','HF','AF','HC','AC','HY','AY','HR','AR','final_result']
new_col = [f"{c}_rolling" for c in colms]
rolling_averages(group, colms, new_col)
```

```python
df_rolling = df.groupby('HomeTeam').apply(lambda x: rolling_averages(x, colms, new_col))
```

```python
df_rolling = df_rolling.droplevel('HomeTeam')
```

```python
df_rolling.index = range(df_rolling.shape[0])
df_rolling
```

```python
def retraing_data(data, predictors):
    train = data.loc[data['Date'] < '2020-01-01']
    test = data[data['Date'] > '2020-01-01']
    rf.fit(train[predictor], train['final_result'])
    prediction = rf.predict(test[predictor])
    combined = pd.DataFrame(dict(actual=test['final_result'], prediction=prediction))
    precision = precision_score(test['final_result'], prediction)
    return combined, precision
```

```python
combined, precision = retraing_data(df_rolling, predictor + new_col)
```
