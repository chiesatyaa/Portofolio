## Data Preparation

```python
df_2011 = pd.read_csv('E0 (1).csv')
df_2012 = pd.read_csv('E0 (2).csv')
df_2013 = pd.read_csv('E0 (3).csv')
df_2014 = pd.read_csv('E0 (4).csv')
df_2015 = pd.read_csv('E0 (5).csv')
df_2016 = pd.read_csv('E0 (6).csv')
df_2017 = pd.read_csv('E0 (7).csv')
df_2018 = pd.read_csv('E0 (8).csv')
df_2019 = pd.read_csv('E0 (9).csv')
df_2020 = pd.read_csv('E0 (10).csv')
df_2021 = pd.read_csv('E0 (11).csv')
df_2022 = pd.read_csv('E0 (12).csv')
df_2023 = pd.read_csv('E0 (13).csv')
df_2024 = pd.read_csv('E0 (14).csv')
```

```python
raw_nan = df[df.isna().any(axis=1)]
raw_nan
```

```python
#Percentage Missing Value of each columns
len_df = len(df)
feature_nan = pd.DataFrame({
    'Miss Count': df.isnull().sum().values,
    'Miss %': 100*(df.isnull().sum().values/len_df)}, index=df.columns)
feature_nan
```

```python
df = df.dropna()
```

```python
df_c = []
for r in df.columns:
    if df[r].dtypes == 'float64':
        df[r] = df[r].astype('int64')
    df_c.append(r)
```

```python
#Delete column by indexing
df = df.drop(columns=df.columns[22:37])
```

```python
df.reset_index(drop=True)
```

## Data Preprocessing

```python
df = df.sort_values(by=['HomeTeam', 'Date'], ascending=True).reset_index()
df
```

```python
df['Home'] = df['HomeTeam'].astype('category').cat.codes
df['Away'] = df['AwayTeam'].astype('category').cat.codes
df
```

```python
for index, col in enumerate(df['FTR']):
    if col == 'H':
        df.at[index, 'FT_Result'] = 3
    elif col == 'D':
        df.at[index, 'FT_Result'] = 1
    elif col == 'A':
        df.at[index, 'FT_Result'] = 0

df['FT_Result'] = df['FT_Result'].astype('int64')
```

```python
df = df.drop(columns=['index'])
```

```python
df[['Home', 'Away']] = df[['Home', 'Away']].astype('int64')
```

```python
df.to_csv('pl_clear.csv')
```

```python
def rolling_averages(team, cols, new_cols):
    team = team.sort_values("Date")
    rolling_stats = team[cols].rolling(3, closed='left').mean()
    team[new_cols] = rolling_stats
    team = team.dropna(subset=new_cols)
    return team
```

```python
col = ['FTHG','FTAG','HTHG','HTAG','HS','AS','HST','AST','FT_Result']
new_col = [f"{c}_rolling" for c in col]

rolling_averages(team,col,new_col)
```

```python
matches_rolling = df.groupby("HomeTeam").apply(lambda x: rolling_averages(x, col, new_col))
matches_rolling
```
